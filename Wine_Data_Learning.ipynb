## 와인 분류 모델

```python
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Input, Dense, Dropout
```

```python
model = Sequential()

model.add(Input(shape=(11,)))

model.add(Dense(100, activation = 'sigmoid'))
model.add(Dropout(0.2))

model.add(Dense(200, activation = 'relu'))
model.add(Dropout(0.4))

model.add(Dense(50, activation = 'tanh'))
model.add(Dropout(0.1))

model.add(Dense(10, activation = 'softmax'))
```

```python
import pandas as pd

# 데이터 불러오기
redwine = pd.read_csv('winequality-red.csv', sep = ';')

# 독립변수와 종속변수 분리
x, y = redwine.iloc[:, :-1], redwine.iloc[:, -1]
```

```python
from sklearn.model_selection import train_test_split

# 훈련 데이터셋과 평가 데이터 셋의로 구분 (여기서 검증은 따로 구분 X)
X_train, X_test, y_train, y_test= train_test_split(x, y)
```

```python
# 훈련 정의
model.compile(loss='sparse_categorical_crossentropy', optimizer='adam', metrics=['accuracy']) # sparse 붙여서 y도 1개 받아도 동작
```

```python
# 데이터 학습
history = model.fit(X_train, y_train, batch_size=200, epochs=300)
```

```python
# 모델 평가
model.evaluate(X_test, y_test)
```
